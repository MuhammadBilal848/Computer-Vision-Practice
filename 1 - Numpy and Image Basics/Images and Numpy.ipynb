```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# keep in mind new version of jupyterlab and jupyter notebook dont require this magic command, we just used it here
# if someone's using older version
```

```python
# to open an image we need to use python imaging library(PIL)
from PIL import Image
# image module helps open up images after which we are going to be able to transfrom this in form of a numpy array
```

```python
one  = Image.open('../Data/angrybird.jpg')
```

```python
one
```

```python
type(one) # here one is an image object as we can see in the output below
# its a jpeg image file and numpy can't actually deal with it because numpy works with arrays right
# so we can transform this image into an array.
```

```python
# asarray is a function in numpy that converts any type of data into an array
one_arr = np.asarray(one) 
type(one_arr) # now its a numpy array
```

```python
one_arr
```

```python
one_arr.shape
# its shape is now (264, 512, 3) 
# height 264
# width 512
# channel 3 (R,G,B)

# image is holding 3 channels R,G,B which means its holding 'single 264 by 512 for red(0)' , 'single 264 by 512 for green(1)'
# 'single 264 by 512 for blue(2)'
```

```python
# plt.imshow() is a function specially designed to show images that transformed to as array
plt.imshow(one_arr) # with the help of this we can easily see the labeling of numbers of height and width
plt.ylabel('Height')
plt.xlabel('Width')
plt.title('Angry Birds')
```

```python
one_red = one_arr.copy() # here we just made copies of one_arr to work with them instead of the real one (they all are same)
one_green = one_arr.copy()
one_blue = one_arr.copy()
```

```python
# now using color channels we can access R G and & B variations of this image
# lets try to access red channel of this image for this we can use copy of image that we have made above
# syntax - array[need every pixel of height or all the rows,need every pixel of width or all the columns,channel index you want]
# Channel - R G B
# Index -   0 1 2

# lets see this image in red channel
# print(one_red[:,:,0]) # red channel values 0-255 [0 means no red pure black, 255 means full pure red]
plt.imshow(one_red[:,:,0])
# we see that the image is not specifically looking red instead giving greenish effect, because matplotlib uses viridis color
# mapping for color blind ppl.
# https://matplotlib.org/stable/gallery/color/colormap_reference.html
```

```python
# lets see this red image in grayscale
plt.imshow(one_red[:,:,0],cmap = 'gray') # closer to white = more red
```

```python
# lets see this image in green channel
# print(one_green[:,:,1]) # green channel values 0-255 [0 means no green pure black, 255 means full pure green]
plt.imshow(one_green[:,:,1])
```

```python
# lets see this red image in grayscale
plt.imshow(one_green[:,:,1],cmap = 'gray') # closer to white = more green
```

```python
# lets see this image in blue channel
# print(one_blue[:,:,2]) # blue channel values 0-255 [0 means no blue pure black, 255 means full pure blue]
plt.imshow(one_blue[:,:,2])
```

```python
# lets see this image in blue channel
plt.imshow(one_blue[:,:,2],cmap='gray') # closer to white = more blue
```

```python
# lets try to zero one of 3 channels
one_blue[:,:,2] = 0 # here we have completely deleted blue channel the image is now having only green and red channel 
plt.imshow(one_blue) 
```

```python
# lets try to remove green channel now
one_blue[:,:,1] = 0 # here we have completely deleted green channel the image is now having only red channel because blue 
# channel has already been deleted from one_blue
plt.imshow(one_blue)
```

```python
print(one_blue.shape)
# here the shape of image is still (264, 512, 3)
# getting confuse how it is still having 3 as channel numbers right?
# deleted means we have set the channel to 0 but not removed it, it's still having channels but with 0 values
# we can check it here as well
plt.imshow(one_blue[:,:,1]) # we have set the values of 1 and 2 channels values to zero
```
