# c::matchTemplate
$$cv.matchTemplate(image, templ, method[, result[, mask]]) ->result$$

#### Following are the methods provided by opencv for template matching
* https://docs.opencv.org/4.x/df/dfb/group__imgproc__object.html#ga586ebfb0a7fb604b35a23d85391329be

```python
# All the 6 methods for comparison in a list
# Note how we are using strings, later on we'll use the eval() function to convert to function
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR','cv2.TM_CCORR_NORMED', 
           'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
# last two methods(4,5) work on finding the min value on the heatmap the rest of the functions find max value.
```

* The result is a heatmat.
* We can see the bright spot is indicating that the face is at that position in the image
* We want to draw a rectangle on the bright spot, so we will use the heatmap's values for rectangle's coordinate.
* We can get heatmap's values using cv2.minMaxLoc that returns minVal,maxVal,minLoc,maxLoc

```python
fin = cv2.matchTemplate(img,tmp,eval(methods[0]))
plt.imshow(fin)
```

```python
fin = cv2.matchTemplate(img,tmp,eval(methods[5]))
plt.imshow(fin)
```

* We will be using for loop for to apply each method using eval()

```python
for m in methods:
    
    # creating copy of img 
    imgc = img.copy()
    
    # using method with eval
    mtd = eval(m)
    
    # applying tempalte matching
    fin = cv.matchTemplate(imgc,tmp,mtd)
    
    # extracting coordinates
    minVal,maxVal,minLoc,maxLoc = cv2.minMaxLoc(fin)
    
    # drawing rectanle
    if m in ['cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']:
        topLeft = minLoc # this looks like this (x,y)
    else:
        topLeft = maxLoc # this looks like this (x,y)
        
    # unpacking the face h ,  w and channels
    height , width , chnl = tmp.shape
    
    # we can assign bottom right corner using the topleft corners topLeft[0]+width will fix the x axis value
    # topLeft[1]+height will fix the y value.
    btmRight = (topLeft[0]+width,topLeft[1]+height)
    
    cv2.rectangle(imgc,topLeft,btmRight,(255,0,0),5)
    
    f = plt.figure(figsize=(10,10))
    f = plt.subplot(121)
    f.imshow(imgc)
    plt.title('Template Detection')

    f = plt.subplot(122)
    f.imshow(fin)
    plt.title(f'Heatmap Using {m}')
    
    
```
