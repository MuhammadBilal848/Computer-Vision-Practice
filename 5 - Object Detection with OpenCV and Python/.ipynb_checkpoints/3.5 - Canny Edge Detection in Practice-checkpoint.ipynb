```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

* Edge detection is nothing but a filtering process.
* There are so many types of filters like Sobel, Roberts, Prewitt and Canny filters available, these filters are based on the first-order derivative.
* The Laplacian of Gaussian is based on the Second-order derivative.
* Some other filters include Morphological Gradient & Laplacian.
* Canny edge detection uses linear filtering with a Gaussian kernel to smooth noise and then computes the edge strength and direction for each pixel in the smoothed image.
* Candidate edge pixels are identified as the pixels that survive a thinning process called non-maximal suppression.
* Canny Edge Detection works in five steps:
    * Apply gaussian filter to smooth the image and remove noise.
    * Find the intensity gradient of the image.
    * Apply non-maximum supression to get rid of spurios response to edge detection
    * Apply double threshold to determine potential edges.
    * Track edge by hysteresis means finalize the detection of edges by suppressing all the other edges that are weak and not connecting to other edges.
    

```python
cat = cv.imread('../Data/Cat 4K Pictures/3.jpg')
cat = cv.cvtColor(cat,cv.COLOR_BGR2RGB)
plt.imshow(cat)
```

```python
cat = cv.medianBlur(cat,3,5)
```

```python
plt.figure(figsize = (10,10))
plt.imshow(cv.Canny(cat,0,255))
```

```python
med_val = np.median(cat)
lower = int(max(0, 0.7* med_val))
upper = int(min(255,1.3 * med_val))
```

```python
plt.figure(figsize = (10,10))
plt.imshow(cv.Canny(cat,lower,upper))
plt.title('Canny Edge Detetion on a Cat')
plt.savefig('Canny Cat Detection')
```

```python
plt.figure(figsize = (10,10))
plt.imshow(cv.Canny(cat,lower,upper+90))
```

```python
plt.figure(figsize = (10,10))
plt.imshow(cv.Canny(cat,50,255))
```
