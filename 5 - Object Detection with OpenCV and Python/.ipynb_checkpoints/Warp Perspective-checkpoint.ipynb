```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

## <center> Perspective view warping is to convert or perspective correction of images from angle to birds eye view transform <center/>

# cv::warpPerspective

* **src** :	Source image. CV_8U , CV_16U , CV_32S , or CV_32F depth and 1, 3, or 4 channels are supported.
* **dst** :	Destination image with the same type as src . The size is dsize .
* **M** :	3x3 transformation matrix.
* **dsize** :	Size of the destination image.

```python
warp = cv.imread('../Data/warp.jpg')
warp = cv.cvtColor(warp,cv.COLOR_BGR2RGB)
plt.imshow(warp)
```

* First step is to find four coordinates of the target you want to warp.
* You can get the coordinates using paint(microsoft built-in).

```python
pt1 = np.int32([[1225,257],[2585,249],[177,3137],[1417,3897]])
```

```python
cwarp = warp.copy()
for (x,y) in pt1:
    print(f'{x} , {y}')
    cv.putText(cwarp,f'{x} , {y}',(x,y),cv.FONT_HERSHEY_COMPLEX,2.5,(255,0,0),5)
    cv.circle(cwarp,(x,y),30,(0,0,255),-1)
plt.figure(figsize= (10,10))
plt.imshow(cwarp)
```

```python
warp.shape
```

* Convert the points to be float.

```python
pt1 = np.float32(pt1)
pt1
```

* Now we set points 2.

```python
width = 300
height = 650
#               first point          second point           third point         fourth point
pt2 = np.float32([[0,0]        ,       [width,0]            ,[0,height]         ,[width,height]])
pt2
```

* Now we will call getPerspectiveTransform function and pass pt1 and pt2.

```python
real_coordinates = cv.getPerspectiveTransform(pt1,pt2)
```

* Now we will call main warpPerspective function.

```python
width,height
```

```python
output = cv.warpPerspective(warp,real_coordinates,(width,height))
```

```python
plt.figure(figsize = (15,15))
plt.imshow(output)
```
