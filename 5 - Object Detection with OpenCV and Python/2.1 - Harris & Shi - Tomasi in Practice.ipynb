```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

* A corner is a point whose local neighborhood stands and two dominant and different as directions.
* A corner can be interpreted as the junction of two edges where an edge is a sudden change in image brightness.
* There are two popular corner detection algorithms:
    * Harris Corner Detection.
    * Shi - Tomasi Corner Detection.
* The basic integration is that the corners can be detected by looking for significant change in all directions.
* Shi - Tomasi works on similar approach as harris corner detection but with different scoring function selection criteria

```python
ubit = cv.imread('../Data/ubit/8.jpg')
cubit = cv.cvtColor(ubit,cv.COLOR_BGR2RGB)
gubit = cv.cvtColor(ubit,cv.COLOR_BGR2GRAY)
```

# <center>CORNOR HARRIS

```python
corubit = cv.cornerHarris(gubit,3,5,0.04)
f = plt.figure(figsize = (10,7))
f = f.add_subplot(111)
f.imshow(corubit,cmap='gray')
```

```python
corubit = cv.dilate(corubit,None)
f = plt.figure(figsize = (10,7))
f = f.add_subplot(111)
f.imshow(cv.dilate(corubit,None))
```

```python
cubit[corubit>0.01*corubit.max()] = [255,0,0]
```

```python
f = plt.figure(figsize = (15,15))
f = f.add_subplot(111)
f.imshow(cubit)
plt.title('Corner Detection Using Harris')
plt.savefig('Ubit Corner Harris')
```

# <center> Shi - Tomasi

```python
ubit = cv.imread('../Data/ubit/8.jpg')
cubit = cv.cvtColor(ubit,cv.COLOR_BGR2RGB)
gubit = cv.cvtColor(ubit,cv.COLOR_BGR2GRAY)
```

```python
corner = np.int0(cv.goodFeaturesToTrack(gubit,50,0.08,8))
```

```python
for cor in corner:
    x,y = cor.ravel()
    cv.circle(cubit,(x,y),5,(0,255,0),-1)
```

```python
f = plt.figure(figsize = (10,10))
f = f.add_subplot(111)
f.imshow(cubit)
```

```python
ubit = cv.imread('../Data/ubit/8.jpg')
cubit = cv.cvtColor(ubit,cv.COLOR_BGR2RGB)
gubit = cv.cvtColor(ubit,cv.COLOR_BGR2GRAY)
corner = np.int0(cv.goodFeaturesToTrack(gubit,130,0.028,11))
print(len(corner))
for cor in corner:
    x,y = cor.ravel()
    cv.circle(cubit,(x,y),4,(0,255,0),-1)
f = plt.figure(figsize = (15,15))
f = f.add_subplot(111)
f.imshow(cubit)
plt.title('Corner Detection Using Shi - Tomasi')
plt.savefig('Ubit Shi Toamsi')
```
